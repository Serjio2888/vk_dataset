```python
import pandas as pd
import users_filters as uf
```

```python
subs = pd.read_csv('./sub/21.02.19-12:21.csv')
subs = subs.replace({pd.np.nan: None})
subs.head()
```

```python
users = pd.read_csv('./users/21.02.19-12:21.csv')
users.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
users = users.replace({pd.np.nan: None})
users.head()
```

```python
print(f"users: {len(users)}, subs: {len(subs)}")
bool(users['bdate'][2])
```

```python
girl = []
num_girl = 0
pop = []
num_pop = 0
pop_history = []
girl_history = []
for num, user in enumerate(users.iterrows()):
    user = user[1]
    filter_girl = []
    filter_pop = []
    if num > -1:
        user_sub = subs[subs['user_id'] == user['user_id']]
        
        filter_girl.append(uf.sex_filter(user, 1))
        filter_girl.append(uf.age_filter(user, list(range(1988, 2001)), True))
        filter_girl.append(uf.top_univer_filter(user, user_sub))
        #filter_girl.append(uf.bank_filter(user, user_sub))
        filter_girl.append(uf.travel_filter(user, user_sub, threshold=0))
        
        filter_pop.append(uf.age_filter(user, list(range(2003, 1950)), True))
        filter_pop.append(uf.nauchpop_filter(user, user_sub, threshold=1))
        
        if all(filter_pop):
            pop.append(user['user_id'])
            num_pop += 1
        pop_history.append(filter_pop)
        if all(filter_girl):
            girl.append(user['user_id'])
            num_girl += 1
        girl_history.append(filter_girl)
        if  num % 500 == 0 and num > 0:
            print(f"girl: {num_girl}, pop: {num_pop}")
            print(f"random_id girl: {girl[-1]}, pop: {pop[-1]}")
            
```

```python
from json import JSONEncoder
j = JSONEncoder()
with open('./filtered_users/21.02.19-12:08_girl.json', 'w') as fw:
    fw.write(j.encode(girl))
with open('./filtered_users/21.02.19-12:08_pop.json', 'w') as fw:
    fw.write(j.encode(pop))
```
